## **KMeans cluster example** 

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

## **Using dataset named as make_blobs**

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

## **Let's plot them first**

```python
plt.figure(figsize=(17, 8))
plt.scatter(X[:, 0], X[:, 1])
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(17, 8))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## **From the above graph, we can clearly observe that from the 4rth point, it aqure the high changing ratio, and than remain constant**
## - So we have 4 clustors (classes) in our dataset

```python
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(17, 8))
plt.scatter(X[:,0], X[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```
